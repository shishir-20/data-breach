```python
import pandas as pd

df = pd.read_csv("../data/raw_data.csv")
df

from sklearn.preprocessing import LabelEncoder

file_type_encoder = LabelEncoder()
df['file_type_encoded'] = file_type_encoder.fit_transform(df['file_type'])

df[['file_type', 'file_type_encoded']]
label_encoder = LabelEncoder()
df['label_encoded'] = label_encoder.fit_transform(df['label'])

df[['label', 'label_encoded']]
```

```python
df_processed = df.drop(['file_name', 'file_type', 'label'], axis=1)
df_processed
```

```python
df_processed.to_csv("../data/processed_data.csv", index=False)
print("✅ processed_data.csv saved")
```
