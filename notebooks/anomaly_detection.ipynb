```python
import pandas as pd

df = pd.read_csv("../data/user_logs.csv")
df
```

```python
X = df.drop("user_id", axis=1)
X
```

```python
from sklearn.ensemble import IsolationForest

anomaly_model = IsolationForest(
    contamination=0.4,   # % of anomalies expected
    random_state=42
)

anomaly_model.fit(X)
print("✅ Anomaly detection model trained")
```

```python
df["anomaly_flag"] = anomaly_model.predict(X)
df
```

```python
import joblib

joblib.dump(anomaly_model, "../model/anomaly_model.pkl")
print("✅ anomaly_model.pkl saved")
```
